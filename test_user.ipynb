### logstic regression 을 적용하기 위해서 categorical data를 integer로 바꾸자
단 이때 원래 칼럼을 바꾸지 말고 catg_ 의 접미사를 붙여서 하자 나중에 문제가 발생한다.

```python
train_users_pd['country_destination'].unique()
```

```python
train_users_pd['gender'] = train_users_pd['gender'].astype('category')
train_users_pd['language'] = train_users_pd['language'].astype('category')
train_users_pd['country_destination'] = train_users_pd['country_destination'].astype('category')
```

```python
print(train_users_pd['country_destination'].cat.categories)
train_users_pd['country_destination'].labels
train_users_pd['country_destination'].describe()
```

```python
train_users_pd.info()
```

```python
cat_columns = train_users_pd.select_dtypes(['category']).columns
cat_columns
```

```python
train_users_pd[cat_columns] = train_users_pd[cat_columns].apply(lambda x: x.cat.codes)
```

```python
train_users_pd[['age','gender','language','country_destination']]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import KFold
```

```python
alg = LogisticRegression()
X_cols = ['age','gender','language']
y_col = 'country_destination'
```

```python
X = train_users_pd[X_cols]
y = train_users_pd[y_col]
```

```python
alg.fit(X,y)
```

```python
alg.predict([[30, 1, 3],[40,0,5]])
```

```python
len(train_users_pd)
```

```python
kf = KFold(len(train_users_pd),n_folds=3)
```

```python
expect_ys = []
for train_indexs, test_indexs in kf:
        train_X = train_users_pd[X_cols].iloc[train_indexs,:]
        train_y = train_users_pd[y_col].iloc[train_indexs]
        alg.fit(train_X, train_y)
        expect_y = alg.predict(train_users_pd[X_cols])
        expect_ys =+ expect_y
```

```python
sum(expect_ys == train_users_pd[y_col]) / len(expect_ys)
```

```python
from sklearn import cross_validation
```

```python
alg = LogisticRegression()
```

```python
scores = cross_validation.cross_val_score(alg,train_users_pd[X_cols], train_users_pd[y_col], cv=3)
```

```python
print(scores.mean())
```
